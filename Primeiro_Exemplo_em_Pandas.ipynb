```python
# Importando biblioteca pandas no Python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Uma forma para ver a base
display(combustiveis_df)
```

```python
# Exibe as 5 primeiras linhas
display(combustiveis_df.head())
```

```python
# Para exibir as primeiras 15 linhas
display(combustiveis_df.head(15))
```

```python
# Comandos Dataframe.shape() e DataFrame.describe() | Shape: mostra linhas e colunas | Describe
print(combustiveis_df.shape)

# Para mostrar somente as linhas:
print(combustiveis_df.shape[0])
```

```python
# Comandos Dataframe.shape() e DataFrame.describe() | Shape: mostra linhas e colunas | Describe: mostra estatísticas "mais básicas" que precisamos e é um metodo usado para resumo estatístico da distribuição de cada coluna numérica,  incluindo contagem, média, desvio padrão, valor minimo, quartis e valor maximo. 
# e se tiver colunas não numericas elas são ignoradas 


print(combustiveis_df.describe())
```

```python
# Quais são as colunas e que tipo de dados elas tem.
print(combustiveis_df.info())
```

```python
# Filtrar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
 # Aqui criamos um dataframe apenas com as colunas desejadas
 ca_df = combustiveis_df[['Produto', 'Revenda', 'Municipio', 'Valor de Venda']]
 display(ca_df)
```

```python
# Exibe a 4º Linha
display(ca_df.loc[3])
```

```python
# Exibir da 10º linha até a 20º linha
display(ca_df.loc[9:19])
```

```python
# Criar dataframe gas_df contendo apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
# gas_df = combustiveis_df[['Produto', 'Revenda', 'Municipio', 'Valor de Venda']]
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Para saber qual posto e município da gasolina mais cara
display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com múltiplas condições para filtragem
#Quais são so preços do ETANOL na minha cidade
etanol_barueri_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'BARUERI')]
etanol_barueri_df.sort_values(by="Valor de Venda")
display(etanol_barueri_df)
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL Brasil todo?
media_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)
```

```python
# Quero adicionar uma coluna de valor booleano no combustiveis_df 
# Chamada "Ativo" que, por padrão, vai ser True para 
# TODAS as linhas
combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())
```
